```python
numerics = ['int64', 'float64']
cat=['object']

dataset_num = dataset.select_dtypes(include = numerics) # has numerical data
dataset_cat = dataset.select_dtypes(include=cat)  # has string data
```

```python
dataset_num = dataset_num.fillna(dataset_num.mean())   #filling empty dataset with mean
```

```python
dataset_cat= dataset_cat.fillna('')  #filling empty dataset with blank spaces
```

```python
dataset_cat.info()
```
